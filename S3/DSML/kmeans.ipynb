```python
from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

```python
iris = datasets.load_iris()
x = iris.data[:, :2]
y = iris.target
```

```python
plt.scatter(x[:,0], x[:,1], c=y, cmap='prism')
plt.xlabel('Sepal Length', fontsize=18)
plt.ylabel('Sepal Width', fontsize=18)
```

```python
km = KMeans(n_clusters=3, init='k-means++', n_init=10, max_iter=300, 
            tol=0.0001, verbose=0, random_state=21, copy_x=True, algorithm='lloyd')  # Use 'lloyd' or 'elkan'
```

```python
km.fit(x)
```

```python
centers = km.cluster_centers_
print("Cluster Centers:\n", centers)
new_labels = km.labels_
print("New Labels:\n", new_labels)
print("True Labels:\n", y)
```
